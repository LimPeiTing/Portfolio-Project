Check for skewness of the dataset 

Skewness is a statistical measure used to describe the shape of a probability distribution.

It measures the asymmetry of a distribution.

Skewness indicates the direction and degree to which the data deviates from a symmetrical bell curve.

A distribution with zero skewness is perfectly symmetrical, meaning the left and right sides of the distribution are mirror images.

Positive skewness or right skewed (skewness > 0) means more weight in the left tail of the distribution, the right tail is longer or fatter than the left, suggesting that the data has a tendency to have higher values.

Negative skewness or left skewed (skewness < 0) means more weight in the right tail of the distribution, the left tail is longer or fatter, implying a tendency towards lower values.

Magnitude of skewness:

(-0.5, 0.5)â€Šâ€”â€”â€Šlow or approximately symmetric.

Beyond -1 and 1â€Šâ€”â€”â€ŠHighly skewed.

elseâ€Šâ€”â€”â€Šmoderately skewed.


```python
from scipy.stats import skew

features = ['age', 'bmi', 'charges']

for col in features:
    s = skew(df[col])
    if s >= -0.5 and s <= 0.5:
        print(f'Skewness of {col} = {s:.4f} [low or approximately symmetric]')
    elif s <= -1:
        print(f'Skewness of {col} = {s:.4f} [highly left skewed]')
    elif s >= 1:
        print(f'Skewness of {col} = {s:.4f} [highly right skewed]')
    else:
        print(f'Skewness of {col} = {s:.4f} [moderately skewed]')
```

```python
# Skewness of age and bmi are both approximately low, followed the normal distribution, which is ideal for the model's learning 
```

Data Encoding: categorical data 

```python
df_encoded = df.copy()

df_encoded['sex'] = df_encoded['sex'].map({'female':0,'male':1})
df_encoded['smoker'] = df_encoded['smoker'].map({'yes':1,'no':0})
df_encoded['region'] = df_encoded['region'].map({'southwest':0,'southeast':1,'northwest':2,'northeast':3})

df_encoded.head()
```

```python
X = df_encoded.drop('charges', axis=1)
y = df_encoded['charges']
```

Data Modelling 

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor 
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
import xgboost as xgb
from sklearn.neighbors import KNeighborsRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Since our target variable has continuous value, hence we will be using regression model for the data modelling
```

Linear regression 

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)

# For regression models, 
lr_r2 = r2_score(y_test, y_pred)
lr_mae = mean_absolute_error(y_test, y_pred)
lr_mse = mean_squared_error(y_test, y_pred)
lr_rmse = np.sqrt(mean_squared_error(y_test, y_pred))
lr_mape = mean_absolute_percentage_error(y_test, y_pred)
```

```python
print(f'R-square (r2):', lr_r2)
print(f'Mean absolute error (MAE):', lr_mae)
print(f'Mean squared error (MSE):', lr_mse)
print(f'Root-mean squared error (RMSE):', lr_rmse)
print(f'Mean absolute percentage error (MAPE):', lr_mape)
```

```python
print(f'R-square (r2):', dt_r2)
print(f'Mean absolute error (MAE):', dt_mae)
print(f'Mean squared error (MSE):', dt_mse)
print(f'Root-mean squared error (RMSE):', dt_rmse)
print(f'Mean absolute percentage error (MAPE):', dt_mape)
```
